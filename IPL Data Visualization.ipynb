# Step 2: Analyze Player's Performance

```python
# Filter data for MS Dhoni
data_1=data[data['striker'] =='DT Christian']
```

```python
# Calculate the number of matches played total run scored and average run per match

mateches_played =len(data_1)
total_runs_scored= sum(data_1['runs_off_bat'])
average_runs_per_match=(total_runs_scored / mateches_played)
```

```python
# Group by 'striker' and calculate the sum of 'runs_off_bat' and count of 'ball'

df0 = pd.DataFrame(data.groupby('striker')['runs_off_bat'].sum()).reset_index()
df1 = pd.DataFrame(data.groupby('striker')['ball'].count()).reset_index()

# Merge the two DataFrames on 'striker' column
df2 = df0.merge(df1, on='striker', how='left')

# Filter for strikers with runs_off_bat >= 500 and ball <= 500
df2 = df2[(df2['runs_off_bat'] >= 500) & (df2['ball'] <= 500)]

df2['strike_rate']= df2['runs_off_bat']/df2['ball']*100
df2.reset_index(inplace=True, drop=True)
```
